```python
# importando pandas para utilizarmos no código.
import pandas as pd
```

```python
# importando o dataset do kaggle que está os dados.
dados = pd.read_csv('C02.csv')
pd.DataFrame(dados)
```

```python
#Função para verificar se há dados nulos ou nao preenchidos.
dados.isna()
```

```python
#Testado uma variável que conta o número de países, excluindo dados que estejam duplicados,assim isolando os dados "países".
country = list(dados['country_name'].drop_duplicates())
country
```

```python
#Realizado o filtro para o dataset ao qual foi realizado o dashboard.

df_br = dados.query("country_name == 'Brazil' & year >= 2001") # variável que contém os dados do Brasil.
df_arg = dados.query("country_name == 'Argentina' & year >= 2001") # variável que contém os dados da Argentina.
df_chi = dados.query("country_name == 'Chile' & year >= 2001") # variável que contém os dados do Chile.
df_countries = pd.concat([df_arg, df_br, df_chi]) # juntando os novos dados em um só dataset
df_countries.update(df_countries['value'].astype('float').round(3)) # diminuindo os valores de emissão para uma melhor leitura visual dos dados.

```

```python
df_countries = df_countries.reset_index(drop = True) # corrigindo a posição dos itens.
df_countries


```

```python
df_countries.to_csv('C02SA.csv', sep = ';') #importando o novo dataset para um arquivo.
```
